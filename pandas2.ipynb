#### Making Changes to our data

```python
import pandas as pd
df = pd.read_csv('./pokemon_data.csv')
```

```python
df.head(5)
```

```python
df["Total"] = (
    df["HP"]
    + df["Attack"]
    + df["Defense"]
    + df["Sp. Atk"]
    + df["Sp. Def"]
    + df["Speed"]
)
df = df.drop(columns=['Total'])
df.head(4)
```

```python
df['Total'] = df.iloc[:, 4:10].sum(axis=1) 
# if axis is set to 0 then it will add vertically
# if axis is set to 1 then it will add horizontally

cols = list(df.columns.values)

df = df[cols[0:4]+ [cols[-1]] + cols[4:12]]
df.head(4)
```
